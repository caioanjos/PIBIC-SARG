```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from geobr import read_state

hospital_counts = tabela_2021.groupby(['SG_UF_NOT', 'HOSPITAL']).size().unstack(fill_value=0)
hospital_counts.columns = ['Internado', 'NÃ£o Internado', 'Ignorado']
estados = read_state(code_state='all', year=2019)
estados_counts = estados.set_index('abbrev_state').join(hospital_counts)
coluna_interesse = 'Internado'  # Pode ser 'NÃ£o Internado' ou 'Ignorado'

fig, ax = plt.subplots(1, 1, figsize=(12, 8))
estados_counts.plot(column=coluna_interesse, cmap='OrRd', linewidth=0.8, ax=ax, edgecolor='0.7', legend=True)
ax.axis('off')  # Remover os eixos
ax.set_title(f'DistribuiÃ§Ã£o de InternaÃ§Ãµes Hospitalares no Brasil em 2021')
plt.show()
```
